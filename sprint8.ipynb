### **Limpieza de datos**

* #### df_viajes_empresas

```python
# Crea una función para limpiar los nombres de las compañías.  
def limpia_nombre(nombre):
    if " - " in nombre: # Verifica si el string contiene el patrón " - ".
        nombre_filtrado = re.search(r"[A-Za-z].*", nombre) # Busca la primera ocurrencia de una letra seguida de cualquier carácter hasta el final.
        return nombre_filtrado.group() # Devuelve la parte encontrada (sin el prefijo antes del patrón).
    return nombre # Si no contiene " - ", devuelve el nombre tal cual.

# Aplica la función limpia_nombre a cada elemento de la columna "company_name".
df_viajes_empresas["company_name"] = df_viajes_empresas["company_name"].apply(limpia_nombre)

# Imprime los valores únicos (ya limpios) de la columna "company_name" de forma ordenada.
print("Valores únicos (limpios) company_name:")
for i, company in enumerate(sorted(df_viajes_empresas["company_name"].unique()), start=1):
    print(f"{i}.- {company}")
```

* #### df_loop_a_ohare

```python
# Elimina los registros en donde la columna "duration_seconds" = 0.0
df_loop_a_ohare = df_loop_a_ohare[df_loop_a_ohare["duration_seconds"] != 0]

# Cambia el tipo de dato de la columna "start_ts" a tipo datetime.
df_loop_a_ohare["start_ts"] = pd.to_datetime(df_loop_a_ohare["start_ts"])

# Convierte los valores de la columna "duration_seconds" a minutos (redondeados a 2 decimales).
df_loop_a_ohare["duration_seconds"] = (df_loop_a_ohare["duration_seconds"] / 60).round(2)

# Cambia el nombre de la columna "duration_seconds" a "duration_minutes".
df_loop_a_ohare.rename(columns={"duration_seconds":"duration_minutes"}, inplace=True)

print(df_loop_a_ohare.head())
```

### **Prueba de hipótesis:**  
* **Hipótesis nula (H₀):**
La duración promedio de los viajes desde el Loop hasta el aeropuerto O'Hare no cambia los sábados lluviosos comparado con los sábados no lluviosos.
* **Hipótesis alternativa (H₁):**
La duración promedio de los viajes sí cambia en sábados lluviosos respecto a sábados no lluviosos.

```python
# Filtra los registros para quedarse solo con los Sábados (día 5).
df_loop_a_ohare_sabados = df_loop_a_ohare[df_loop_a_ohare["start_ts"].dt.dayofweek == 5]

# Filtra Sábados con clima lluvioso ("Bad").
sabados_lluviosos = df_loop_a_ohare_sabados[df_loop_a_ohare_sabados["weather_conditions"] == "Bad"]

# Filtra Sábados con clima no lluvioso ("Good").
sabados_no_lluviosos = df_loop_a_ohare_sabados[df_loop_a_ohare_sabados["weather_conditions"] == "Good"]

# Calcula el promedio de duración en minutos para Sábados lluviosos.
promedio_dias_lluviosos = sabados_lluviosos["duration_minutes"].mean()

# Calcula el promedio de duración en minutos para Sábados no lluviosos.
promedio_dias_no_lluviosos = sabados_no_lluviosos["duration_minutes"].mean()

# Imprime los promedios.
print(f"Promedio duración sábados lluviosos: {promedio_dias_lluviosos:.2f} minutos")
print(f"Promedio duración sábados no lluviosos: {promedio_dias_no_lluviosos:.2f} minutos")

# Realiza la prueba de Levene para comparar varianzas.
stat_levene, p_levene = stats.levene(
    sabados_lluviosos["duration_minutes"],
    sabados_no_lluviosos["duration_minutes"]
)

print(f"\nResultado prueba de Levene:")
print(f"Statistic: {stat_levene:.3f}")
print(f"P-value: {p_levene:.4f}")

# Evalúa el resultado de la prueba de Levene.
if p_levene > 0.05:
    print("No se rechaza la hipótesis nula: se asume igualdad de varianzas.")
    equal_var = True
else:
    print("Se rechaza la hipótesis nula: no se asume igualdad de varianzas.")
    equal_var = False

# Realiza la prueba t para muestras independientes usando el resultado de Levene.
t_stat, p_value = stats.ttest_ind(
    sabados_lluviosos["duration_minutes"],
    sabados_no_lluviosos["duration_minutes"],
    equal_var=equal_var,
    nan_policy="omit"
)

print(f"\nResultado prueba t:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretación final de la prueba t.
if p_value < 0.05:
    print("La diferencia en duración promedio es estadísticamente significativa.")
else:
    print("No hay evidencia estadística para afirmar que la duración promedio cambia.")
```
